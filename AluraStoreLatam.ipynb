### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# 1. Calculamos el ingreso total sumando la columna 'Precio'
ingreso_t1 = tienda1['Precio'].sum()
ingreso_t2 = tienda2['Precio'].sum()
ingreso_t3 = tienda3['Precio'].sum()
ingreso_t4 = tienda4['Precio'].sum()

# 32. Imprimimos los resultados para comparar
print(f"Ingreso Total Tienda 1: ${ingreso_t1}")
print(f"Ingreso Total Tienda 2: ${ingreso_t2}")
print(f"Ingreso Total Tienda 3: ${ingreso_t3}")
print(f"Ingreso Total Tienda 4: ${ingreso_t4}")
```

```python
import matplotlib.pyplot as plt

# Datos calculados previamente
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [tienda1['Precio'].sum(), tienda2['Precio'].sum(), tienda3['Precio'].sum(), tienda4['Precio'].sum()]

plt.figure(figsize=(10, 6))
plt.bar(tiendas, ingresos, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])
plt.title('Ingresos Totales por Tienda', fontsize=15)
plt.ylabel('Ingresos (en miles de millones)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 2. Ventas por categoría

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Combinamos todas las tiendas en un solo DataFrame
all_stores = pd.concat([
    tienda1.assign(Tienda='Tienda 1'),
    tienda2.assign(Tienda='Tienda 2'),
    tienda3.assign(Tienda='Tienda 3'),
    tienda4.assign(Tienda='Tienda 4')
])

# 1. Gráfico de Dispersión: Concentración de Ventas
plt.figure(figsize=(10, 7))
sns.scatterplot(data=all_stores, x='lon', y='lat', hue='Tienda', alpha=0.5)
plt.title('Mapa de Dispersión de Ventas por Ubicación')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.grid(True)
plt.show()

# 2. Mapa de Calor de Calificaciones (Satisfacción por Región)
plt.figure(figsize=(10, 7))
scatter = plt.scatter(all_stores['lon'], all_stores['lat'],
                      c=all_stores['Calificación'], cmap='RdYlGn', alpha=0.6)
plt.colorbar(scatter, label='Calificación')
plt.title('Mapa de Satisfacción del Cliente por Ubicación')
plt.show()
```
